## Merging and Concatenation

```python
# Read datasets
courses = pd.read_csv("courses.csv")
nov = pd.read_csv("reg-month1.csv")
dec = pd.read_csv("reg-month2.csv")
students = pd.read_csv("students.csv")
```

```python
# Concat 'nov' and 'dec' data and name it 'registration' or 'reg'
reg = pd.concat([nov,dec], ignore_index=True)
reg.head(2)
```

```python
# Multi-indexing
pd.concat([nov, dec], keys = ['nov', 'dec'])
```

```python
# inner join (reg and students)
reg.merge(students, on="student_id", how="inner")
```

```python
# right join
reg.merge(students, on="student_id", how="right")
```

```python
# left join
reg.merge(students, on="student_id", how="left")
```

```python
# outer join
reg.merge(students, on="student_id", how="outer")
```

```python
# Find total revenue generated
reg.merge(courses, on="course_id")['price'].sum()
```

```python
# Find month by month revenue (Advanced)
pd.concat([nov,dec], keys = ['nov', 'dec']).reset_index().merge(courses, on = "course_id").groupby("level_0")['price'].sum()
```

```python
# Print the registration table 
# cols -> name -> course -> price

reg.merge(courses, on = "course_id").merge(students, on="student_id")[['name', 'course_name', 'price']]
```

```python
# Plot bar chart for revenue/course (revenue per course)
course = reg.merge(courses, on = "course_id").groupby("course_name")
course["price"].sum().plot(kind="bar")
```

```python
# 5. Print student name -> partner name for all enrolled students
# self join
students.merge(students, left_on="partner", right_on="student_id")
```

```python
# Find top 3 students who did most number enrollments
reg.merge(students, on="student_id").groupby(['student_id', 'name']).size().sort_values(ascending=False).head(3)
```

```python
# Find top 3 students who spent most amount of money on courses
reg.merge(students, on="student_id").merge(courses, on="course_id").groupby(['student_id', 'name'])['price'].sum().sort_values(ascending=False).head(3)
```
