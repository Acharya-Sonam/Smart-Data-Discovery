#### *i) Display the bar chart of top 5 teams who have played maximum number of matches in the IPL.*
`Hint:` Be careful the data is divided in 2 different cols(Team 1 and Team 2)

```python
# Code here 
total_matches = df['Team1'].value_counts().add(df['Team2'].value_counts(),
                                               fill_value=0)
top_5_teams = total_matches.sort_values(ascending=False).head(5)

print(top_5_teams)
```

#### *ii) Player who got Most no. of player of the match award playing against Mumbai Indians.*

```python
# Code here 
mi_matches = df[(df['Team1'] == 'Mumbai Indians') | (df['Team2'] == 'Mumbai Indians')]
against_MI = mi_matches[mi_matches['WinningTeam'] != 'Mumbai Indians']['Player_of_Match']
top_player_against_mi = against_MI.value_counts().head(1)

print("Player with most PoM awards against Mumbai Indians:",top_player_against_mi)



```

#### *v) Team1 vs Team2 Dashbord* 

*Create a function which will take two string(name of two teams) as input. Show win Loss record between them <br>
and player getting most player of the match award in matches between these two teams.* <br>
`team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')`

```python
def team1_vs_team2(t1, t2):
    mask = ((df['Team1'] == t1) & (df['Team2'] == t2)) | \
           ((df['Team1'] == t2) & (df['Team2'] == t1))
    
    rivalry = df[mask]
    print(f"Win Loss Record between {t1} and {t2}")
    print(rivalry['WinningTeam'].value_counts())
    print(f"\nPlayer with most POTM in these matches")
    print(rivalry['Player_of_Match'].value_counts().head(1))

# Execution
team1_vs_team2('Kolkata Knight Riders', 'Chennai Super Kings')
```
